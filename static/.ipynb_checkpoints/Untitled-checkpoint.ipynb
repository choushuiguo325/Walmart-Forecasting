```python
import pandas as pd
```

```python
food_path = 'sales_train_evaluation_2.csv'
calendar_path = 'calendar.csv'
food_df = pd.read_csv(food_path)
calendar_df = pd.read_csv(calendar_path)
```

```python
food_df.head()
```

```python
calendar_df.head()
```

```python
# calendar_df["week_count"] = calendar_df["wm_yr_wk"] - 11100
```

```python
week_count = calendar_df['wm_yr_wk']
```

```python
year = calendar_df["year"]
month = calendar_df["month"]
```

```python
food_3090 = food_df.loc[0]
# number = length(food_3090)
food_3090 = food_3090[1:].to_list()
food_3090
```

```python
len(food_3090)
```

```python
len(week_count)
```

```python
food_dict = {"volume": food_3090[:1941], "week": week_count[:1941]}
food_info = pd.DataFrame(food_dict)
food_info
```

```python
list(range(1,278))
```

```python
food_data = food_info.groupby(["week"]).mean()
food_sum = food_info.groupby(["week"]).sum()["volume"]
food_data["week_count"] = list(range(1,279))
food_data["weekly_sum"] = food_sum
food_data = food_data.rename(columns={"volume":"weekly_mean"})
food_data = food_data.reset_index(drop= True)
```

```python
food_data = food_data[["week_count", "weekly_mean", "weekly_sum"]]
food_data = food_data.reset_index(drop = True)
food_data
```

```python
food_data.to_csv("weekly_data.csv", index = False)
```

```python
monthly_dict = {"volume": food_3090[:1941], "month": month[:1941], "year": year[:1941]}
```

```python
monthly_data = pd.DataFrame(monthly_dict)
monthly_data = monthly_data.groupby(["year", "month"]).mean()
monthly_data = monthly_data.reset_index(drop = True)
monthly_data = monthly_data.rename(columns = {"volume": "monthly_mean"})
monthly_data['month_count'] = list(range(1,66))
monthly_data = monthly_data[["month_count", "monthly_mean"]]
monthly_data
```

```python
monthly_data.to_csv("monthly_data.csv", index = False)
```
